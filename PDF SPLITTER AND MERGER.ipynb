```python
from pypdf import PdfReader, PdfWriter
import os

def merge_pdfs(pdf_list, output_path):
    pdf_writer = PdfWriter()

    for pdf in pdf_list:
        pdf_reader = PdfReader(pdf)
        for page in pdf_reader.pages:
            pdf_writer.add_page(page)

    with open(output_path, 'wb') as out_file:
        pdf_writer.write(out_file)

    print(f"✅ Merged PDF saved as '{output_path}'")


def split_pdf(input_pdf, output_folder):
    pdf_reader = PdfReader(input_pdf)

    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for i, page in enumerate(pdf_reader.pages):
        pdf_writer = PdfWriter()
        pdf_writer.add_page(page)

        output_path = os.path.join(output_folder, f"page_{i+1}.pdf")
        with open(output_path, 'wb') as out_file:
            pdf_writer.write(out_file)

        print(f"📄 Saved: {output_path}")


# ---------- Example Usage ----------

if __name__ == "__main__":
    # 🟩 Merge Example
    pdfs_to_merge = ['2201020793 SEM RESULTS.pdf', 'AADHAR SUMIT.pdf']  # Replace with your actual files
    merge_pdfs(pdfs_to_merge, 'merged_output.pdf')

    # 🟥 Split Example
    pdf_to_split = 'PPT OS CASE STUDY GROUP1.pdf'  # Replace with your actual file
    split_output_dir = 'split_pages'
    split_pdf(pdf_to_split, split_output_dir)
```
